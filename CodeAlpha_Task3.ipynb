***GLOBAL LAYOFFS DATA VISUALIZATION DASHBOARD***

IMPORTNG LIBRARIES -

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

LOADING DATASET -

```python
import zipfile
with zipfile.ZipFile('/content/Layoffs.zip', 'r') as zip_ref:
    zip_ref.extractall('/content')
```

```python
df = pd.read_csv('/content/Layoffs.zip')
```

**DATA VISUALIZATION**

```python
#1--Top 10 Companies by Total Layoffs :

top_companies = (
    df.groupby('company')['total_laid_off']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,6))
top_companies.plot(kind='barh')
plt.title("Top 10 Companies by Total Layoffs")
plt.xlabel("Total Employees Laid Off")
plt.ylabel("Company")
plt.tight_layout()
plt.show()
```

### Interpretation

- This visualization shows the top 10 companies with the highest total layoffs.
- Companies at the top experienced the largest workforce reductions.
- The chart highlights organizations that were most affected during the global layoff period.

```python
#2--Top 10 Countries by Total Layoffs

top_countries = (
    df.groupby('country')['total_laid_off']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,6))
top_countries.plot(kind='bar')
plt.title("Top 10 Countries by Total Layoffs")
plt.xlabel("Country")
plt.ylabel("Total Employees Laid Off")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Interpretation

- This chart compares the total number of employees laid off across the top 10 countries.
- It highlights the countries that experienced the highest workforce reductions.
- The visualization helps identify the geographical distribution of layoffs.

```python
#3--Top 10 Industries by Total Layoffs

top_industries = (
    df.groupby('industry')['total_laid_off']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,6))
top_industries.plot(kind='bar', color='teal')
plt.title("Top 10 Industries by Total Layoffs")
plt.xlabel("Industry")
plt.ylabel("Total Employees Laid Off")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Interpretation

- This visualization identifies the industries with the highest number of layoffs.
- It provides insights into which sectors were most affected during the observed period.
- Such information can help understand industry-specific employment trends.

```python
#4--Layoffs Trend Over Time

# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')
# Create a new 'year' column
df['year'] = df['date'].dt.year
# Group data by year and calculate total layoffs
yearly_layoffs = df.groupby('year')['total_laid_off'].sum()

plt.figure(figsize=(10,6))
plt.plot(
    yearly_layoffs.index,
    yearly_layoffs.values,
    marker='o',
    linewidth=2,
    color='blue'
)

plt.title("Total Layoffs by Year", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Employees Laid Off", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(yearly_layoffs.index)
plt.tight_layout()
plt.show()
```

### Interpretation

- The line chart illustrates the yearly trend in global layoffs.
- It highlights periods where workforce reductions increased or decreased.
- The visualization helps identify years with the highest layoff activity and overall employment trends over time.

```python
#5--Industry Distribution

industry_counts = df['industry'].value_counts().head(8)
plt.figure(figsize=(8,8))
plt.pie(
    industry_counts,
    labels=industry_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'edgecolor': 'white'}
)

plt.title("Top 8 Industries Distribution")
plt.tight_layout()
plt.show()
```

### Interpretation

- The pie chart shows the distribution of the top industries in the dataset.
- It highlights which industries contribute the largest share of layoff records.
- This visualization provides a quick understanding of industry representation.

```python
#6--Correlation Heatmap

plt.figure(figsize=(8,6))

correlation = df[['total_laid_off',
                  'percentage_laid_off',
                  'funds_raised']].corr()
sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)

plt.title("Correlation Between Numerical Features")
plt.tight_layout()
plt.show()
```

### Interpretation

- The heatmap displays the correlation between numerical variables.
- Positive values indicate a direct relationship, while negative values indicate an inverse relationship.
- This helps identify whether funding, layoffs, and layoff percentages are related.

**INSIGHTS OF GLOBAL LAYOFFS DATA VISUALIZATION DASHBOARD**

- The highest number of layoffs was concentrated in a few major companies.
- The United States experienced the largest number of reported layoffs.
- Certain industries were affected more heavily than others.
- Layoff activity showed noticeable fluctuations over different years.
